# Getting Historical Data (OHLC & Volume) - Part-1

```python
from binance.client import Client
import pandas as pd
```

Connecting to the account on Binance.com

```python
keys = pd.read_csv("~/Documents/temp/bnc/api_key.txt", sep=" ", header=None)
api_key = keys[0][0]
secret_key = keys[0][1]
client = Client(api_key = api_key, api_secret = secret_key, tld = "com")
```

Binance started in 2017 so the earliest historical data available is from 2017

```python
timestamp = client._get_earliest_valid_timestamp(symbol = "BTCUSDT", interval = "1d")
timestamp # earliest data available on Binance
```

```python
pd.to_datetime(timestamp, unit = "ms") # earliest data available on Binance
```

```python
bars = client.get_historical_klines(symbol = "BTCUSDT",
                                   interval = "1d", start_str = timestamp, limit = 1000)
bars # returned as JSON data
```

```python
df = pd.DataFrame(bars) # converting JSON to data frame
df
```

```python
df["Date"] = pd.to_datetime(df.iloc[:,0], unit = "ms") # adds a Date column to the end of the table
df
```

```python
df.columns = ["Open Time", "Open", "High", "Low", "Close",
              "Volume", "Close Time", "Quote Asset Volume",
              "Number of Trades", "Taker Buy Base Asset Volume",
              "Taker Buy Quote Asset Volume", "Ignore", "Date"]
df
```

```python
df = df[["Date", "Open", "High", "Low", "Close", "Volume"]].copy()
df
```

```python
# setting the Date column as the index
df.set_index("Date", inplace = True)
df
```

```python
df.info()
```

```python
for column in df.columns:
    df[column] = pd.to_numeric(df[column], errors = "coerce")
df
```
