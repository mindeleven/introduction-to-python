Matplotlib is a Python 2D plotting library which produces publication quality figures in a variety of hardcops formats (png, jpg, etc.) and displays plots in enviroments across platforms. Examples can be found at https://matplotlib.org/2.0.2/gallery.html

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
# simple example with two numpy arrays
x = np.linspace(0,5,11) # linearly spaced
y = x ** 2
```

```python
x
```

```python
y
```

```python
# two basic ways to create matplotlib plots
# (1) functional, (2) object oriented
# (1) a functional way to create matplotlib plots
# Signature: plt.plot(*args, scalex=True, scaley=True, data=None, **kwargs)
# Docstring:
# Plot y versus x as lines and/or markers.
# 
# Call signatures::
# 
#     plot([x], y, [fmt], *, data=None, **kwargs)
#     plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)
# 
# The coordinates of the points or line nodes are given by *x*, *y*.
plt.plot(x,y)
```

```python
plt.plot(x,y,'r-') # adding line color to edit the plot
# adding x and y label
plt.xlabel('X label')
plt.ylabel('y label')
# adding title
plt.title('Title')
```

```python
# creating multi plots on the same canvas
plt.subplot(1,2,1) # num of rows, num of columns, number of the plot you're referring to
plt.plot(x,y,'r')

plt.subplot(1,2,2) # adding second plot
plt.plot(y,x,'b')
```

```python
# (2) object oriented method
fig = plt.figure() # creates a figure object, a imaginary blank canvas
# adding axes to this figure by passing a list
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, width, height
# plot on the axes
axes.plot(x,y) # same plot like earlier with object oriented method
# putting x-label, y-label and title
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('X Y Title')
```
